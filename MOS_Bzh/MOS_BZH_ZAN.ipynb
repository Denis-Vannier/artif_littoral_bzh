```python
import pandas as pd
import geopandas as gpd

mos=gpd.read_file("MOS_Bretagne_2011_2021/MOS_bretagne.shp")
```

```python
##### Filtrage des parcelles artificialisées en 2021  ########
conso_agri=['Bâtiment agricole','Serre','Carrière']
artif2021=mos[(mos["type"]=="Consommé") | (mos["lib4_2021"].isin(conso_agri))]

##### Filtrage des parcelles artificialisées entre 2011 et 2021 ########
# Changement du classement des parcelles "consommées",
# avec l'ajout des bâtiments agricoles, des serres et des carrières :
artif=['Autre infrastructure', 'Habitat individuel','Jardins habitat (unité foncière)', 'Chemin-sentier','Voie desserte mixte', 
       'Parc et jardin','Autre équipement local, administration', 'Voie desserte habitat','Terrain vacant - habitat', 
       'Sport et loisir', 'Habitat collectif',
       'Hôtel', 'Voie desserte activité', "Equipement d'enseignement",
       'Jardins activité (unité foncière)', 'Bâti divers',
       'Activité tertiaire', 'Urbain mixte (habitat/activité tertiaire)',
       'Activité autre que tertiaire : industrie et artisanat',
       'Equipement sportif (construit)', 'Cimetière',
       'Surface commerciale', 'Terrain vacant - autre',
       'Equipement pour eau, assainissement, énergie',
       'Terrain vacant - activité', 'Equipement de santé',
       'Parking non construit', 'Activités multiples',
       'Activités autres que tertiaires : logistique et stockage',
       'Infrastructure de transport', 'Parking construit',
       'Terrain vacants infrastructures (chantiers)',
       'Espaces "naturels" militaires', 'Bâtiment remarquable',
       'Espaces naturels et agricoles aéroports', 'Défense (anthropisé)','Bâtiment agricole','Serre','Carrière']
nature=["Terre agricole","Espaces boisés","Espace naturel","Terrain vacant - agricole","Réseau hydrographique","Plan d'eau", "Rocher et falaise", 'Plage, dune et sable']


artif2021.loc[(artif2021["lib4_2011"].isin(nature)) & (artif2021["lib4_2021"].isin(artif)), "Conso_2021"]="oui"
artif2021["Conso_2021"]=artif2021["Conso_2021"].fillna("non")
artif2021=artif2021[['code_insee', 'nom_commun', 'code_siren', 'epci', 'Conso_2021', 'surface_m2', 'lib4_2021','lib4_2011','tx_bati','geometry','type']]

artif2021.info()
```

```python
##### Filtrage des parcelles artificialisées dans une commune de la bande littorale des 10km  ########

# Chargement des polygones des bandes littorales :
Bande10kmComLitt=gpd.read_file("BANDE_LITTORALE/BANDE_LITTORALE_SECTEURS_COM_BZH.shp")
Bande10kmComLitt=Bande10kmComLitt.to_crs("EPSG:2154")
litt=Bande10kmComLitt["INSEE_COM"].tolist()
artif2021["code_insee"]=artif2021["code_insee"].astype("str")
artif2021_litt=artif2021[artif2021["code_insee"].isin(litt)]

artif2021_litt.to_file("MOS_Bretagne_2011_2021/MOS_Bzh_Litt_Conso_2011_2021.shp")

agri=['Bâtiment agricole','Serre']
autres_activites=['Activité tertiaire', 'Activité autre que tertiaire : industrie et artisanat','Surface commerciale','Activités autres que tertiaires : logistique et stockage']

artif2021_litt.loc[artif2021_litt["Conso_2021"]=="oui", "Conso_tot_m2"]=artif2021_litt["surface_m2"]
artif2021_litt["Conso_tot_m2"]=artif2021_litt["Conso_tot_m2"].fillna(0)

artif2021_litt.loc[(artif2021_litt["lib4_2021"].isin(agri)) & (artif2021_litt["Conso_2021"]=="oui"), "Conso_BatiAgric_m2"]=artif2021_litt["surface_m2"]
artif2021_litt["Conso_BatiAgric_m2"]=artif2021_litt["Conso_BatiAgric_m2"].fillna(0)
artif2021_litt.loc[(artif2021_litt["lib4_2021"].isin(autres_activites)) & (artif2021_litt["Conso_2021"]=="oui"), "Conso_InduComTerti_m2"]=artif2021_litt["surface_m2"]
artif2021_litt["Conso_InduComTerti_m2"]=artif2021_litt["Conso_InduComTerti_m2"].fillna(0)

######### Bilan par communes (comprises dans la bande littorale 10km) ############
artif2021_litt_Bilan=artif2021_litt.groupby(["code_insee","nom_commun","epci","code_siren"], as_index=False).agg(
   Artif_Tot_m2=("surface_m2","sum"),
   Conso_Tot_m2=("Conso_tot_m2","sum"),
   Conso_bati_agric_m2=("Conso_InduComTerti_m2","sum"),
   Conso_InduComTerti_m2=("Conso_BatiAgric_m2","sum"))
# Calcul de la part des bâtiments agricoles et des serres dans les parcelles consommées entre 2011 et 2021 :
artif2021_litt_Bilan["Part_Conso_bati_agric"]=( artif2021_litt_Bilan["Conso_bati_agric_m2"] / artif2021_litt_Bilan["Conso_Tot_m2"] ) * 100
artif2021_litt_Bilan["Part_Conso_bati_agric"]=artif2021_litt_Bilan["Part_Conso_bati_agric"].round(1)
# Calcul de la part des parcelles consommées entre 2011 et 2021 sur l'ensemble des terres artificialisées en 2021 :
artif2021_litt_Bilan["Part_Conso_2011-2021"]=( artif2021_litt_Bilan["Conso_Tot_m2"] / artif2021_litt_Bilan["Artif_Tot_m2"] ) * 100
artif2021_litt_Bilan["Part_Conso_2011-2021"]=artif2021_litt_Bilan["Part_Conso_2011-2021"].round(1)
artif2021_litt_Bilan=artif2021_litt_Bilan.sort_values(by="Part_Conso_2011-2021", ascending=False)
artif2021_litt_Bilan.to_csv("MOS_Bretagne_2011_2021/BilanCom_Artif_COM_MOS_Bzh_2011-2021.csv")

######### Bilan global selon l'origine et la destination des parcelles consommées ############
# (de toutes les communes comprises dans la bande littorale 10km) 
artif2021_litt_conso=artif2021_litt[artif2021_litt["Conso_2021"]=="oui"]
origine_destination=artif2021_litt_conso.groupby(["lib4_2011","lib4_2021"], as_index=False).agg(
   surface_m2=("surface_m2","sum"))
origine_destination["surface_m2"]=origine_destination["surface_m2"].round(1)
origine_destination.to_csv("MOS_Bretagne_2011_2021/BilanCom_MOS_Bzh_Origine_destinantion_2011-2021.csv")
```
